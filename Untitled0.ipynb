```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, roc_curve, auc, confusion_matrix
from google.colab import drive

# 1. CONECTAR AL DRIVE
drive.mount('/content/drive')

# 2. RUTA DEL ARCHIVO (Ajusta el nombre si lo cambiaste en tu Drive)
ruta = '/content/drive/MyDrive/Proyecto_Bullying/dataset_bullying_anonimo.xlsx'

# 3. LEER LOS DATOS
df = pd.read_excel(ruta)
print(f"✅ ¡Conexión exitosa! Analizando {len(df)} registros anonimizados.\n")

# 4. PROCESAMIENTO
le = LabelEncoder()
for col in ['Grado', 'Zona_Junin']:
    df[col] = le.fit_transform(df[col].astype(str))

# Eliminamos ID_Alumno para que la IA no se confunda, solo usaremos las variables psicológicas
X = df.drop(columns=['Alerta_Bullying', 'ID_Alumno'])
y = df['Alerta_Bullying']

# 5. ENTRENAMIENTO (Test al 30% para que no sea 100% y sea más realista)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modelo con Poda (Pruning) para que los gráficos salgan con curvas y variedad
clf = DecisionTreeClassifier(max_depth=4, min_samples_leaf=5, random_state=42)
clf.fit(X_train, y_train)

# 6. MÉTRICAS
y_pred = clf.predict(X_test)
y_probs = clf.predict_proba(X_test)[:, 1]
print(f"📊 Precisión Final: {accuracy_score(y_test, y_pred) * 100:.2f}%")

# =============================================================
# BLOQUE DE LOS 5 GRÁFICOS (PORTADA DE GITHUB / INFORME)
# =============================================================
plt.style.use('seaborn-v0_8-whitegrid')

# --- GRÁFICO 1: MATRIZ DE CONFUSIÓN ---
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='g', cmap='YlGnBu')
plt.title('1. Matriz de Confusión: Validación de Alertas', fontsize=14)
plt.xlabel('Predicción de la IA')
plt.ylabel('Realidad')
plt.show()

# --- GRÁFICO 2: IMPORTANCIA DE VARIABLES (Multivariable) ---
plt.figure(figsize=(10, 6))
importancias = pd.Series(clf.feature_importances_, index=X.columns)
importancias.sort_values().plot(kind='barh', color='teal')
plt.title('2. Importancia de Factores (¿Qué detecta la IA?)', fontsize=14)
plt.show()

# --- GRÁFICO 3: MAPA DE CORRELACIÓN (Heatmap) ---
plt.figure(figsize=(12, 10))
sns.heatmap(df.drop(columns=['ID_Alumno']).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('3. Mapa de Correlación: Relación entre Variables', fontsize=14)
plt.show()

# --- GRÁFICO 4: CURVA ROC (Validación Estadística) ---
fpr, tpr, _ = roc_curve(y_test, y_probs)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=3, label=f'Curva ROC (Área = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.title('4. Curva ROC: Capacidad del Modelo', fontsize=14)
plt.legend(loc="lower right")
plt.show()

# --- GRÁFICO 5: EL ÁRBOL DE DECISIÓN VISUAL ---
plt.figure(figsize=(20, 10))
plot_tree(clf,
          feature_names=X.columns,
          class_names=['Estable', 'Riesgo'],
          filled=True,
          rounded=True,
          fontsize=10)
plt.title("5. Lógica del Árbol de Decisión Final", fontsize=18)
plt.show()
```
